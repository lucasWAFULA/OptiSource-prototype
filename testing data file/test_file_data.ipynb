```python
import numpy as np
import pandas as pd

np.random.seed(42)

n_sources = 300

risk_levels = np.random.choice(
    ["low", "medium", "high"],
    size=n_sources,
    p=[0.4, 0.35, 0.25]
)

records = []

for risk in risk_levels:

    if risk == "low":
        deception_score = np.random.uniform(0.05, 0.25)
        reliability_score = np.random.uniform(0.7, 0.9)
        ci_flag = np.random.binomial(1, 0.05)
        handling_cost_kes = np.random.uniform(20_000, 80_000)
        handler_confidence = np.random.uniform(0.7, 0.9)
        report_accuracy = np.random.uniform(0.7, 0.9)
        task_success_rate = np.random.uniform(0.7, 0.9)

    elif risk == "medium":
        deception_score = np.random.uniform(0.25, 0.6)
        reliability_score = np.random.uniform(0.45, 0.7)
        ci_flag = np.random.binomial(1, 0.25)
        handling_cost_kes = np.random.uniform(60_000, 150_000)
        handler_confidence = np.random.uniform(0.5, 0.75)
        report_accuracy = np.random.uniform(0.5, 0.75)
        task_success_rate = np.random.uniform(0.5, 0.75)

    else:  # high risk
        deception_score = np.random.uniform(0.6, 0.95)
        reliability_score = np.random.uniform(0.2, 0.45)
        ci_flag = np.random.binomial(1, 0.6)
        handling_cost_kes = np.random.uniform(120_000, 300_000)
        handler_confidence = np.random.uniform(0.3, 0.6)
        report_accuracy = np.random.uniform(0.3, 0.6)
        task_success_rate = np.random.uniform(0.25, 0.5)

    corroboration_score = np.random.uniform(0.3, 0.9)
    report_timeliness = np.random.uniform(0.4, 0.9)
    report_frequency = np.random.randint(1, 12)
    access_level = np.random.randint(1, 6)
    information_value = np.random.uniform(0.3, 1.0)
    threat_relevant_features = np.random.randint(1, 10)
    scenario_probability = np.random.uniform(0.2, 0.9)

    records.append({
        "risk_level": risk,
        "task_success_rate": task_success_rate,
        "corroboration_score": corroboration_score,
        "report_timeliness": report_timeliness,
        "handler_confidence": handler_confidence,
        "deception_score": deception_score,
        "ci_flag": ci_flag,
        "report_accuracy": report_accuracy,
        "report_frequency": report_frequency,
        "access_level": access_level,
        "information_value": information_value,
        "handling_cost_kes": handling_cost_kes,
        "threat_relevant_features": threat_relevant_features,
        "reliability_score": reliability_score,
        "scenario_probability": scenario_probability
    })

df = pd.DataFrame(records)
df.to_csv("synthetic_sources_risk_levels.csv", index=False)

df.head()
```
